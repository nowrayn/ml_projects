```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Nischal PR/NISCHAL NOWRAY\ML/projects jupyter notebooks/Nischal Data Set/House_price_prediction/train.csv")
```

```python
df
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
im_df = df[['YearBuilt', 'TotalBsmtSF', 'TotRmsAbvGrd', 'GarageArea',
            'PoolArea', 'Neighborhood', 'Heating', 'KitchenQual']]
```

```python
Y_train = df['SalePrice']
```

```python
im_df
```

```python
sel_num_col = im_df.select_dtypes(include=['int64','float64']).columns.tolist()
sel_cat_col = im_df.select_dtypes(include=['object','category']).columns.tolist()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder


num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median'))
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', num_pipeline, sel_num_col),
    ('cat', cat_pipeline, sel_cat_col)
])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

full_pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('model', model)
])
```

```python
full_pipeline.fit(im_df, Y_train)
```

```python
import joblib

joblib.dump(full_pipeline, "model_pipeline.pkl")
```
